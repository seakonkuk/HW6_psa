```python
from random import *
import csv
import matplotlib.pyplot as plt

if __name__ == "__main__":
    f = open('random.csv', 'w')

    ran = [100, 1000, 10000, 100000]

    for r in ran:
        for i in range(r):
            num = randint(1, 6)
            f.write(str(num) + '\n')
    f.close()

    ydata = []

    f1 = open('random.csv')

    data = csv.reader(f1)

    for row in data:
        ydata.append(int(row[0]))

    f1.close()

    plt.title('dice frequency histogram')
    plt.xlabel('dice number')
    plt.ylabel('frequency')
    plt.hist(ydata)
    plt.show()

    print("*** 주사위 시뮬레이션 결과 ***")
    print("주사위 시뮬레이션 결과로 가능한 숫자 : 1, 2, 3, 4, 5, 6")
    print("시행 횟수 : 100, 1000, 10000, 100000 번 시행")
    print("분석 : 1 ~ 6의 숫자에 대해서 매우 비슷한 빈도수가 발생했음을 알 수 있다.")
```
