```python
# 7 ~ 9시 사이에서
# 막대 1 : 최대 승차역 30개
# 막대 2 : 최대 하차역 30개
# 막대 3 : 최대 승하차역 30개

# 인덱스 - 데이터
# 3 : 역이름
# 10 : 승차 1
# 12 : 승차 2
# 11 : 하차 1
# 13 : 하차 2


import csv
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns
%matplotlib inline

if __name__ == '__main__':
    f = open('subway.csv')
    data = csv.reader(f)
    next(data)
    next(data)
    
    result1 = {}  # 막대 1 : 최대 승차역 30개
    result2 = {}  # 막대 2 : 최대 하차역 30개
    result3 = {}  # 막대 3 : 최대 승하차역 30개
    
    for row in data:
        if row[3]!='' and row[10]!='' and row[11]!='' and row[12]!='' and row[13]!='':
            result1[row[3]] = int(row[10].replace(',', '')) + int(row[12].replace(',', ''))
            result2[row[3]] = int(row[11].replace(',', '')) + int(row[13].replace(',', ''))
            result3[row[3]] = int(row[10].replace(',', '')) + int(row[12].replace(',', '')) + int(row[11].replace(',', '')) + int(row[13].replace(',', ''))
            
    result1 = sorted(result1.items(), key=lambda x:x[1], reverse = True)
    result2 = sorted(result2.items(), key=lambda x:x[1], reverse = True)
    result3 = sorted(result3.items(), key=lambda x:x[1], reverse = True)
        
    x1 = []
    y1 = []
    x2 = []
    y2 = []
    x3 = []
    y3 = []
    
    for i in range(0, 30):
        x1.append(result1[i][0])
        y1.append(result1[i][1])
        
        x2.append(result2[i][0])
        y2.append(result2[i][1])
        
        x3.append(result3[i][0])
        y3.append(result3[i][1])
        
    rc('font', family='AppleGothic')

    plt.rcParams['axes.unicode_minus'] = False
                
    plt.title('2023-03 평균 최대 승차역 30개')
    plt.xlabel('a subway station')
    plt.ylabel('the number of passengers')
    plt.barh(x1, y1)
    plt.legend()
    plt.show()
    
    plt.title('2023-03 평균 최대 하차역 30개')
    plt.xlabel('a subway station')
    plt.ylabel('the number of passengers')
    plt.barh(x2, y2)
    plt.legend()
    plt.show()
    
    plt.title('2023-03 평균 최대 승히차역 30개')
    plt.xlabel('a subway station')
    plt.ylabel('the number of passengers')
    plt.barh(x3, y3)
    plt.legend()
    plt.show()
    
    print("*** 2023년 3월 교통카드 통계자료를 이용한 출근 시간대 사람들이 가장 많이 타고 내리는 역 분석 결과 ***")
    print("2023년 3월 최대 승차역 30개 중 가장 승차객이 많은 역 : 구로디지털단지역")
    print("2023년 3월 최대 하차역 30개 중 가장 하차객이 많은 역 : 역삼역")
    print("2023년 3월 최대 승하차역 30개 중 가장 승하차객이 많은 역 : 구로디지털단지역")
    print("본 분석 주제가 가장 많이 타고 내리는 역이기 때문에 분석 대상은 승차객 + 하차객이 평균적으로 많은 역으로 결론내려야 한다.")
    print("따라서 승하차 평균 가장 많은 구로디지털단지가 출근 시간대(7 ~ 9시)에 가장 많다고 결론 내릴 수 있다.")
```
